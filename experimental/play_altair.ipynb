```python
import numpy as np
import altair as alt
import pandas as pd
import math
```

```python
x = np.logspace(1 + math.log10(2), 2 + math.log10(2), 100)
```

```python
y1 = [xi for xi in x]
y2 = [xi - math.sqrt(xi) for xi in x]
```

```python
df = pd.DataFrame({"x": x, "y1": y1, "y2": y2})
```

```python
df
```

```python
g1 = alt.Chart(df).mark_line().encode(x="x", y=alt.Y("y1", scale=alt.Scale(type="log", base=10, nice=False)))
g2 = (
    alt.Chart(df)
    .mark_line()
    .encode(
        x="x",
        y=alt.Y(
            "y2",
            scale=alt.Scale(type="log", base=10, nice=False),
            axis=alt.Axis(
                grid=True,
                title="ouhoua",
                titleAnchor="middle",
                tickCount=5,
                labelExpr="datum.value > 50 ? null : datum.label",
            ),
        ),
    )
)
```

```python
l = alt.layer(g1, g2)
l | l.resolve_scale(y="independent")
```
