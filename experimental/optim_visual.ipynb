# Visualise results from optim algo for peqs

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("../results_iter.csv")
```

```python
df.head()
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("iter:N"), alt.Y("loss:Q"), alt.Color("speaker_name")
)
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("iter:N"), alt.Y("score:Q"), alt.Color("speaker_name")
)
```

```python
df_score = df.groupby("speaker_name").agg({"score": ["min", "max"]}).score
df_score.keys()
```
